```python
import qiskit
from qiskit.circuit import QuantumCircuit, Parameter
```

```python
num_layers = 2
num_qubits = 3

qc = QuantumCircuit(num_qubits)
for l in range(num_layers):
    for q in range(num_qubits):
        qc.u(Parameter(f"p{l}{q}{0}"),Parameter(f"p{l}{q}{1}"),Parameter(f"p{l}{q}{2}"), q)
    for q in range(num_qubits):
        qc.cx(q,(q+1)%num_qubits)

qc.draw("mpl", filename="example_ansatz.png")
```
